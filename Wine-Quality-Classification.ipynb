* but we did it without feature scaling , lets do some feature scaling then try fiting these algos again

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
```

```python
X_train_minmax = min_max_scaler.fit_transform(X_train)
X_test_minmax = min_max_scaler.transform(X_test)
```

```python
X_train_standard  = standard_scaler.fit_transform(X_train)
X_test_standard = standard_scaler.transform(X_test)
```

```python
models = {
    "Logistic Regression": LogisticRegression(solver='liblinear'),
    "Decision Tree Classifier": DecisionTreeClassifier(),
    "Random Forest Classifier": RandomForestClassifier(n_estimators=100)
}
```

```python
def evaluate_models(X_train , X_test , y_train ,y_test):
    result = {}
    for name , model in models.items():
        model.fit(X_train , y_train)
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test , y_pred)
        result[name] = accuracy
    return result
```

```python
result_minmax = evaluate_models(X_train_minmax ,X_test_minmax , y_train , y_test)
```

```python
result_minmax
```

```python
result_standard  = evaluate_models(X_train_standard , X_test_minmax , y_train , y_test)
```
